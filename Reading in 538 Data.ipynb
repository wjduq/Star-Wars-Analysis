```python
import pandas as pd
filename = 'star-wars-survey_StarWars.csv'   # file needs to be in working directory
file = open(filename, 'r')

def getColumns(file):
    file.seek(0)
    columns = file.readline().strip().split(',')
        
    return columns


def buildLol(file):
    '''Build a list of lists representation of table.'''        
    lol = []
    for line in file:
        curRow = line.strip().split(',')
        lol.append(curRow)
        
    return lol

columns = getColumns(file)
StarWarsLol = buildLol(file)
five38df = pd.DataFrame(StarWarsLol)
five38df[:3]
```
